```python
#Step 1.2 - Implementing a Bag of Words from scratch by 4 steps

import string
from collections import Counter
import pprint
documents = ['Hello, how are you!',
             'Win money, win from home.',
             'Call me now.',
             'Hello, Call hello you tomorrow?']

lower_case_documents:list[str] = []
for sentence in documents:
    lower_case_documents.append(sentence.lower())

sans_punctuation_documents:list[str]= []
for sentence in lower_case_documents:
    clean_text = sentence.translate(str.maketrans('','',string.punctuation)).lower()
    sans_punctuation_documents.append(clean_text)

preprocessed_documents:list[list[str]] = []
for sentence in sans_punctuation_documents:
    preprocessed_documents.append(sentence.split())

frequency_list = []
for sentence in preprocessed_documents:
    frequency_list.append(dict(Counter(sentence)) )
    
pprint.pprint(frequency_list)

```

```python
#Implementing a Bag of Word with Skitlearn

from sklearn.feature_extraction.text import CountVectorizer

count_vector = CountVectorizer(lowercase=True, stop_words='english')

documents = ['Hello, how are you!',
                'Win money, win from home.',
                'Call me now.',
                'Hello, Call hello you tomorrow?']

pprint.pprint(documents)

# Essas duas linhas fazem a mesma coisa que #1
# count_vector.fit(documents)
# count_vector.get_feature_names_out()

# 1
count_vector.fit_transform(documents)

doc_array = count_vector.transform(documents).toarray()

frequency_matrix = pd.DataFrame(doc_array)

list_names = count_vector.get_feature_names_out().tolist()

frequency_matrix.columns = list_names


frequency_matrix
# print(count_vector.get_feature_names_out())
# print(doc_array)
```
