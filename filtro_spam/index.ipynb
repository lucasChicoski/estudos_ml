```python
#Step 1.2 - Implementing a Bag of Words from scratch by 4 steps

import string
from collections import Counter
import pprint
documents = ['Hello, how are you!',
             'Win money, win from home.',
             'Call me now.',
             'Hello, Call hello you tomorrow?']

lower_case_documents:list[str] = []
for sentence in documents:
    lower_case_documents.append(sentence.lower())

sans_punctuation_documents:list[str]= []
for sentence in lower_case_documents:
    clean_text = sentence.translate(str.maketrans('','',string.punctuation)).lower()
    sans_punctuation_documents.append(clean_text)

preprocessed_documents:list[list[str]] = []
for sentence in sans_punctuation_documents:
    preprocessed_documents.append(sentence.split())

frequency_list = []
for sentence in preprocessed_documents:
    frequency_list.append(dict(Counter(sentence)) )
    
pprint.pprint(frequency_list)

```
